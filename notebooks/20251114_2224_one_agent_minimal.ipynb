Single agent making one API call to generate a business idea.

```python
import anthropic

api_key = "YOUR_API_KEY"  # Get from console.anthropic.com
client = anthropic.Anthropic(api_key=api_key)

print('Client ready!')
```

```python
# Configuration
TOPIC = "AI + Climate Change"
```

```python
# Single agent call
response = client.messages.create(
    model='claude-sonnet-4-5-20250929',
    max_tokens=1000,
    messages=[{
        'role': 'user',
        'content': f'Generate a creative business idea combining: {TOPIC}'
    }]
)

result = response.content[0].text
print(result)
```
