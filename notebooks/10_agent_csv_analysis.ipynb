```python
import anthropic
import pandas as pd

api_key = "YOUR_API_KEY"  # Get from console.anthropic.com
client = anthropic.Anthropic(api_key=api_key)

print('Client ready!')
```

```python
# Generate CSV with 10 topics (1 sentence each)
data = pd.DataFrame({
    'topic': [
        'AI system that predicts patient readmission risk',
        'Blockchain platform for academic credential verification',
        'IoT sensors for precision agriculture monitoring',
        'ML model for real-time fraud detection in banking',
        'NLP chatbot for mental health support',
        'Computer vision for quality control in manufacturing',
        'Recommendation engine for personalized learning',
        'Predictive maintenance for industrial equipment',
        'Sentiment analysis for brand reputation tracking',
        'Autonomous drone delivery for rural healthcare'
    ]
})
data.to_csv('topics.csv', index=False)
print('topics.csv created with 10 topics')
```

```python
# Load CSV and pass ALL topics to agent
df = pd.read_csv('topics.csv')
all_topics = df['topic'].tolist()
topics_text = '\n'.join(f'{i+1}. {t}' for i, t in enumerate(all_topics))
print(f'Topics:\n{topics_text}')

response = client.messages.create(
    model='claude-sonnet-4-5-20250929',
    max_tokens=2000,
    messages=[{
        'role': 'user',
        'content': f'Here are 10 business ideas:\n{topics_text}\n\nRank them by market potential and explain your top 3.'
    }]
)

print(response.content[0].text)
```
