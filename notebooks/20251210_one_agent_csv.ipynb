```python
import anthropic
import pandas as pd

api_key = "YOUR_API_KEY"
client = anthropic.Anthropic(api_key=api_key)

print('Client ready!')
```

```python
# Generate sample CSV
data = pd.DataFrame({
    'topic': ['AI + Healthcare', 'Blockchain + Education', 'IoT + Agriculture']
})
data.to_csv('topics.csv', index=False)
print('topics.csv created')
```

```python
# Load CSV
df = pd.read_csv('topics.csv')
topic = df['topic'].iloc[0]
print(f'Topic: {topic}')
```

```python
# Single agent call
response = client.messages.create(
    model='claude-sonnet-4-5-20250929',
    max_tokens=1000,
    messages=[{
        'role': 'user',
        'content': f'Generate a creative business idea combining: {topic}'
    }]
)

print(response.content[0].text)
```
